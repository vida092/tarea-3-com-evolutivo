```python
import pandas as pd
import numpy as np
```

```python
real1 = pd.read_csv("undx_1_ejecuciones.csv")
```

```python
real1
```

```python
columna = real1['mejores_aptitudes']

media = columna.mean()
desviacion_estandar = columna.std()
mediana = columna.median()
minimo = columna.min()
maximo = columna.max()

print(f"Media: {media}")
print(f"Desviación Estándar: {desviacion_estandar}")
print(f"Mediana: {mediana}")
print(f"Mínimo: {minimo}")
print(f"Máximo: {maximo}")
```
